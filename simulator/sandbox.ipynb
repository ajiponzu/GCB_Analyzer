```python
import numpy as np
import matplotlib.pyplot as plt

# パラメータ
f = 1000  # 1kHz
d_values = np.linspace(0, 2 * np.pi, 2)  # 0から2πまで8ステップ
t_values = np.linspace(0, 200 * np.pi, 10000)  # 時間0から200πまで10000ステップ

# フーリエ変換結果の複素平面上でのプロット（色を明確に変化）
plt.figure(figsize=(10, 6))
colors = plt.cm.viridis(np.linspace(0, 1, len(d_values)))

for i, d in enumerate(d_values):
    A = np.sin(2 * np.pi * f * t_values + d)
    A_fft = np.fft.fft(A)
    plt.plot(np.real(A_fft), np.imag(A_fft), label=f'd = {d:.2f}', color=colors[i])

plt.title('A(t) のフーリエ変換結果（複素平面上）')
plt.xlabel('実部')
plt.ylabel('虚部')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# サンプル信号を生成（ここではサイン波）
fs = 1000  # サンプリング周波数 (Hz)
t = np.arange(0, 1, 1/fs)  # 1秒間の時間軸を生成
signal = np.sin(2 * np.pi * 10 * t) + np.sin(2 * np.pi * 20 * t)  # 10Hzと20Hzのサイン波を合成

# フーリエ変換を実行
fft_result = np.fft.fft(signal)

# 周波数軸を生成
frequencies = np.fft.fftfreq(len(signal), 1/fs)

# 実部、虚部、絶対値、位相を取得
real_part = np.real(fft_result)
imaginary_part = np.imag(fft_result)
absolute_value = np.abs(fft_result)
phase_angle = np.angle(fft_result)  # 位相を取得

# グラフを描画
plt.figure(figsize=(12, 8))

plt.subplot(411)
plt.plot(frequencies, real_part)
plt.title('Real Part')

plt.subplot(412)
plt.plot(frequencies, imaginary_part)
plt.title('Imaginary Part')

plt.subplot(413)
plt.plot(frequencies, absolute_value)
plt.title('Absolute Value')

plt.subplot(414)
plt.plot(frequencies, phase_angle)
plt.title('Phase Angle (radians)')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_digital_signal(period, duration,dT):
    """
    周期的に0と1が入れ替わるディジタル信号波形を生成する関数

    Parameters:
    - period: 信号の周期
    - duration: 生成する信号の継続時間

    Returns:
    - t: 時間軸
    - signal: ディジタル信号波形
    """
    t = np.linspace(0, duration, int(duration / dT))
    signal = np.zeros_like(t)

    cT = 0
    for i in range(len(t)):
        p = int(cT/period)
        if p % 2 == 1:
            signal[i] = 1
        cT += dT

    return t, signal

#---------------------------------------------
# パラメータ設定
period = 1/64  # 周期
duration = 1.0  # 信号の継続時間
dT = period/100

# ディジタル信号波形生成
t, signal = generate_digital_signal(period, duration,dT)

# フーリエ変換を実行
fft_result = np.fft.fft(signal)

# 周波数軸を生成
frequencies = np.fft.fftfreq(len(signal), 1/fs)

# 中央部分を拡大
start_freq = 0  # 開始周波数
end_freq = 20  # 終了周波数
start_index = np.where(frequencies == start_freq)[0][0]
end_index = np.where(frequencies == end_freq)[0][0]

# 絶対値を取得
absolute_value = np.abs(fft_result)

# 位相を取得
phase_angle = np.angle(fft_result)


# グラフを描画
plt.figure(figsize=(20, 6))

plt.subplot(211)
plt.plot(frequencies[start_index:end_index], absolute_value[start_index:end_index])
plt.title('Absolute Value (Frequency Domain)')
plt.xlabel('Frequency (Hz)')

plt.subplot(212)
plt.plot(frequencies[start_index:end_index], phase_angle[start_index:end_index])
plt.title('Phase Angle (Frequency Domain)')
plt.xlabel('Frequency (Hz)')

plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
T=1.0

# サンプル数
N = 1024

# サンプリング周波数
Fs = N / T  # Tはサンプル全体の時間

# 信号生成
t = np.linspace(0, T, N)
sinsignal = np.sin(2 * np.pi * 100 * t - np.pi)  # 周波数10 Hzのサイン波
condition = sinsignal > 0.5  # 条件（2より大きい）に合致する要素をブール型の配列で取得
signal = (condition.astype(float) - 0.5)*2

# FFTを計算
fft_result = np.fft.fft(signal)

# 周波数軸の生成
frequencies = np.fft.fftfreq(N, 1 / Fs)


magnitude = np.abs(fft_result)

# 絶対値が最大のインデックスを求める
magnitude[0] = 0
peak_index = np.argmax(magnitude)
print(f'peak_index={peak_index}')

# ピークの周波数を計算
frequencies = np.fft.fftfreq(len(signal), 1 / Fs)  # 周波数軸
peak_frequency = frequencies[peak_index]

# ピーク周波数における位相（angle）を計算
peak_angle = np.angle(fft_result[peak_index]) * 180 / np.pi

print("ピーク周波数: {} Hz".format(peak_frequency))
print("ピーク周波数における位相: {} degree".format(round(peak_angle,2)))


# プロット
plt.plot(frequencies[:N//2], np.abs(fft_result[:N//2]))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('FFT Result')
plt.grid(True)
plt.show()
```
